```python
import re #regular expressions are sequence of character that help to match or find string in another string
import os
import csv #classes to read and write tabular data in csv
c = []
b = ["ip","timestamp","timezone","request","slash","http","status"]
with open('logfile.txt') as f:
  lines = f.read().splitlines()
  for line in lines:
    line = line.replace(",","")
    line = line.replace("[","")
    line = line.replace("]","")
    line = line.replace('"','')
    line = line.replace('-','')
    a = re.split('\s+',line)[:7]
    c.append(a)
```

```python
len(c)
```

```python
c[0]
```

```python
import csv
with open('logfile.csv','w') as f:
  #csv.writer to get the content into the file
  write = csv.writer(f)
  #writerrow(),writerows() write data to csv file
  write.writerow(b)
  write.writerows(c)
```
